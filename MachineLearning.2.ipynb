```

from collections import Counter
grades=[83,95,91,87,70,0,85,82,100,67,73,77,0]

histogram=Counter(min(grade//10*10,90)for grade in grades)

plt.bar([x+5 for x in histogram.keys()],
        histogram.values(),
        10,
        edgecolor=(0,0,0))

plt.axis([-5,105,0,5])

plt.xticks([10* i for i in range(11)])
plt.xlabel("Decile")
plt.ylabel("# of Students")
plt.title("Distrubution of Exam 1 Grades")
plt.show()
```

```
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [ x+y for x , y in zip(variance,bias_squared)]
xs = [i for i , _ in enumerate(variance)]


plt.plot(xs,variance, 'g-' , label='variance')
plt.plot(xs,bias_squared, 'r-.' , label='bias^2')
plt.plot(xs,total_error, 'b:' , label='total error')


plt.legend(loc=9)
plt.xlabel("model complexity")
#plt.xticks([])
plt.title("The Bias-Variance Tradeoff")
plt.show()
```

```
import matplotlib.pyplot as plt

friends = [70, 65, 72, 63, 71, 64, 60, 64, 67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']

plt.scatter(friends, minutes)


for label, friend_count, minute_count in zip(labels, friends, minutes):
    plt.annotate(label,
                 xy=(friend_count, minute_count),
                 xytext=(5, -5),
                 textcoords='offset points')

plt.title("Daily Minutes vs. Number of Friends")
plt.xlabel("# of friends")
plt.ylabel("Daily minutes spent on the site")


plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'scores':[56,67,45,89,90,72,78,88,94,60,73,81,85,68,77]
})

data['scores'].hist(bins=8,edgecolor='black')


plt.title("Histogram of Exam Scores")
plt.xlabel("Score")
plt.ylabel("Number of Students")
plt.show()
```

```
import seaborn as sns

sns.histplot(data['scores'], bins=8, kde=True)
plt.title("Histogram of Exam Scores with KDE Curve")
plt.show()
```

```
import matplotlib.pyplot as plt

class_A =[56,67,45,89,90,72,78,88,94,60]
class_B =[66,70,65,80,85,79,90,88,84,73]
class_C =[50,52,47,60,65,70,75,78,72,68]


plt.boxplot([class_A,class_B,class_C], tick_labels=["Class A","Class B","Class C"])


plt.title("Boxplot of Exam Scores By Class")
plt.xlabel("Class")
plt.ylabel("Score")

plt.show()
```

```
import seaborn as sns
import matplotlib.pyplot as plt

data=[
    [88,92,80,89,100],
    [78,85,70,83,90],
    [90,91,85,87,93],
    [65,70,72,68,75],
    [95,98,96,92,99]
]

subjects=["Math","Science","English","History","Art"]
students=["Alice","Bob","Charlie","David","Emma"]

sns.heatmap(data,annot=True,fmt="d",
            xticklabels=subjects,yticklabels=students,
            cmap="YlGnBu")


plt.title("Student Exam Scores Heatmap")

plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2*np.pi,200)
plt.figure()

plt.plot(x,np.sin(x),label="sin(x)")
plt.plot(x,np.cos(x),label="cos(x)")
plt.title("Sine and Cosine Waves")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

```
data=np.random.normal(loc=0,scale=1,size=1000)

plt.figure()
plt.hist(data,bins=30,edgecolor="black")
plt.title("Normal(0,1) Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
